```python
import math

i = 0
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for c, s in results[:20]:
    ages, ahe, aft, zhe, zft = c
    row = math.floor(i / 3)
    col = i % 3
    axs[row, col].set_title(f"Sample {i}")
    axs[row, col].plot(ages, ahe)
    axs[row, col].plot(ages, aft)
    axs[row, col].plot(ages, zhe)
    axs[row, col].plot(ages, zft)
    axs[row, col].legend(SampleGenerator.TEMPS)
    axs[row, col].set_xlabel('Age (Ma)')
    axs[row, col].set_ylabel('Elevation (Km)')
    i += 1
```
